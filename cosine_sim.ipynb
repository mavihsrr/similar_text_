```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
file = 'Normalaization of terms 21 Sep 2023.xlsx'
df = pd.read_excel(file)
terms = df['Terms'].tolist()
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(terms)
```

```python
cosinesim = cosine_similarity(tfidf_matrix, tfidf_matrix)
similarityThreshold = 0.88
```

```python
similarPairs = []
```

```python
for i in range(len(terms)):
    term1 = terms[i]
    for j in range(i + 1, len(terms)):
        term2 = terms[j]
        similarity = cosinesim[i][j]
        if similarity >= similarityThreshold:
            similarPairs.append((term1, term2, similarity))
```

```python
for pair in similarPairs:
    term1, term2, similarity = pair
    print("New Data:", term1)
    print("Similar Data:", term2)
    print("Similarity Score:", similarity)
    print()
```

### Asking Input from user 


```python
userin=input("Enter the term for which you would like to find the most revelant search ---  ").strip().lower()

mostSimData = None
maxSim = 0.0
for term in terms:
    similarity= cosine_similarity(vectorizer.transform([userin]), vectorizer.transform([term]))[0][0]
    if similarity > maxSim:
        maxSim= similarity
        mostSimData= term
if mostSimData:
    print(f"Relevant search found is '{mostSimData}' (similarity score of {maxSim:.2f})")
else:
    print(f"Ay no revelant search was found")




```
